```python
import csv

keywords = ["97", "vapormax", "kobe", "ronaldo", "airmax", "90", "jordan"]

reviews_dict = {}
for keyword in keywords:
    reviews_dict[keyword] = []
    
with open('../data.csv', 'r') as csvfile:
    reader = csv.DictReader(csvfile, delimiter=',')
    for row in reader:
        review = {
            'Review': row['comments']
        }
        for keyword in keywords:
            if keyword in review['Review'].lower():
                reviews_dict[keyword].append(review['Review'])

# print(reviews_dict)
```
